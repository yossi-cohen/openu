```python
import pandas as pd
import numpy as np
import random as rd
from sklearn.decomposition import PCA
from sklearn import preprocessing # for scaling the data
import matplotlib.pyplot as plt
```

# PCA 1

To demonstrate the operation of PCA, lets create a set of random numbers scattered linearily 
<br>($y=1.5x+b$) 

```python
x = 10 * np.random.rand(200)
y = 1.5 * x + np.random.randn(200) * 2
plt.scatter(x, y);
plt.axis('equal');
```

Create a PCA instance from sklearn with 2 principle components 
<br>and feed it with the data above

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)

# we take the transpose, as we are fitting among features of the data set.
pca.fit(np.array([x,y]).T)

print('each pca.component is a vector explaining the distribution of the data')
print(pca.components_)
print()
print('the variance for each pca component')
print(pca.explained_variance_)
```

## plot the PCA components

to understand the meaning of the numbers, lets examine them visually
<br> we plot the PCA components as vectors where their direction is given by PC and their length by the explained variance
<br>The length of the PCs describe how important is this PC for describing the data.

```python
plt.scatter(x, y, alpha=0.2)
for length, vector in zip(pca.explained_variance_, pca.components_):
    v = vector * 2 * np.sqrt(length)
    plt.annotate('', pca.mean_ + v, pca.mean_, 
                arrowprops = dict(arrowstyle='->', linewidth=3))
plt.axis('equal');
```

## Dimentionality Reduction

```python
pca = PCA(n_components=1)
pca = pca.fit(np.array([x,y]).T)
x_transformed = pca.transform(np.array([x,y]).T)

plt.axis('equal');
plt.scatter(x, y)

# Transform data back to its original space.
x_transformed = pca.inverse_transform(x_transformed)
plt.scatter(x_transformed[:, 0], x_transformed[:, 1], alpha=0.8)
```

# PCA 2 (Iris)

## Load the data

```python
df = pd.read_csv('data/iris.csv', sep=',', header=None)
df = df.drop(df.index[-1])
classes = df.iloc[:, -1]
features = df.iloc[:, 0:4]
df.head()
```

## perform PCA (4D -> 2D)

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

features_norm = StandardScaler().fit_transform(features)
pca = PCA(n_components = 2)
principalComponents = pca.fit_transform(features_norm)
df_pca = pd.DataFrame(data = principalComponents, columns = ['PC 1', 'PC 2'])
df_pca['class'] = classes
df_pca.head()
```

## plot data in PCA space

```python
fig = plt.figure(figsize = (8,8))
ax = fig.add_subplot(1,1,1) 
ax.set_xlabel('PC 1')
ax.set_ylabel('PC 2')
ax.set_title('2D PCA')
targets = ['Iris-setosa', 'Iris-versicolor', 'Iris-virginica']
for target in targets:
    indicesToKeep = df_pca['class'] == target
    ax.scatter(df_pca.loc[indicesToKeep, 'PC 1'],
               df_pca.loc[indicesToKeep, 'PC 2'])
ax.legend(targets)
ax.grid()
```

# Example (Iris)

## read the csv data set into a data frame

```python
df = pd.read_csv('data/iris.csv', sep=',', header=None)
df.head()
```

As we can see in the results of top 10 rows of the data set, there are 4 columns (features of the data set) and one class column (last column). 
<br>We will reduce the dimensionality of this data set from 4D to 2D.

## Extract the last column (class label)
We will extract the last column of the data set, and store it in a new data frame.
<br>We'll have two data framesone for features and another one for class label.

```python
class_label = pd.DataFrame(df.iloc[:, -1])
class_label.columns = ['label']
df = df.iloc[:, :-1]
df.head()
```

## Normalizing the data
This is one of the most important steps in PCA 
<br>as it is possible that different variables are measured in different scales.
<br>This makes sure that your end result is not dominated by a single variable.

### Mean normalization
Mean normalization is implemented by taking mean of every column, and then subtracting the mean vector from every row/record.

```python
df = df.sub(df.mean(axis=0), axis=1)
df.head()
```

## Calculating covariance
The formula for covariance is: <font size="4">$S = \frac{1}{n} XX^T$</font>
<br>In the above formula, $X$ is the data matrix and $n$ is the number of rows in the data set.
<br>numpy has a built-in function for covariance.

```python
# we will first transform our features data frame into a matrix, 
# and then calculate covariance on feature matrix.
df_mat = np.asmatrix(df)
sigma = np.cov(df_mat.T)
```

We have taken df_mat.T, as we are finding covariance among features of the data set. 
<br>sigma will be our resultant covariance matrix with 4X4 dimensions.

## Finding eigen values and eigen vectors
We found the relationships among the features in the previous step.
<br>Now, we'll find the eigen values and eigen vectors of the covariance matrix sigma. 
<br>This is also called an eigen decomposition. 
<br>The eigen values tell us the variance in the data set and eigen vectors tell us the corresponding direction of the variance. 
<br>The following code snippet finds the eigen values and eigen vectors 

```python
eigVals, eigVec = np.linalg.eig(sigma)
```

From the above formula, we will get 4 eigen values and the coresponding eigen vectors matrix of 4X4 dimensions.

```python
print('eigen values')
print(eigVals)
print()
print('eigen vectors')
print(eigVec)
```

## Sort the eigen values and eigen vectors
As we are interested in maximum variance across features, 
<br>we will sort the eigen values in decreasing order and retrieve their corresponding eigen vectors accordingly.

```python
sorted_index = eigVals.argsort()[::-1] 
eigVals = eigVals[sorted_index]
eigVec = eigVec[:,sorted_index]
```

## Select top k eigen values and corresponding eigen vectors
To reduce dimensions of the data set from d features to k features, 
<br>we select the top k eigen values and the corresponding eigen vectors 
<br>(in our case k=2, as we are reducing data set to 2 dimensions). 

```python
eigVec = eigVec[:,:2]
print('shape:', eigVec.shape, '(only 2 features)')
```

## Forming the new data set in reduced dimensions
We now have the top 2 eigen vectors and the original matrix
<br>PCA being a linear technique, forms linear equation between the old data set and new data set.
<font size="3">$newdataset = dotproduct( [olddataset], [eigen vector] )$</font>

```python
transformed = df_mat.dot(eigVec)
print('transformed shape:', transformed.shape)
```

Let us attach the class label to this new data set and form a new data frame which will be ready to plot.

```python
# horizontally stack transformed data set with class label.
final_df = np.hstack((transformed, class_label))

# convert the numpy array to data frame
final_df = pd.DataFrame(final_df)

# define the column names
final_df.columns = ['x','y','label']
```

The above code makes the final_df data frame of the dimensions: $150 X 3$, 
<br>which has three columns:$x, y, label$

## Plot the new data set

```python
groups = final_df.groupby('label')

figure, axes = plt.subplots()
axes.margins(0.05)

for name, group in groups:
    axes.plot(group.x, group.y, marker='o', linestyle='', ms=6, label=name)

axes.set_title('PCA on iris.data')
axes.legend()
plt.xlabel("principal component 1")
plt.ylabel("principal component 2")
plt.show()
```
